# Satırları Sütun Değerlerine Göre Filtreleme

```python
movies3=pd.read_csv('http://bit.ly/imdbratings')
```

```python
sayaç=0#For döngüsünde normal bir dizi gibi kullanabiliyoruz...
boolens=[]
for length in movies3.duration:
    if length>=160:
        sayaç=sayaç+1
        boolens.append(False)
    else:
        boolens.append(True)
print("Sayaç="+str(sayaç))
```

```python
karşılaştırma=movies3.duration >=200#Bu yazımla pandas kütüphanesinin bize sağladığı kolaylıkla çok daha basit bir şekilde filtreleme yapabiliriz...
movies3[karşılaştırma]
```

```python
movies3[movies3.duration>=200]#Bu şekilde daha kısa bir yazımı da tercih edebiliriz...
#Ancak bu karşılaştırmada sadece görüntüleme elde edebiliriz filtrelemeyi ilerkisatırlarda yada başka bir yerde kullanamayız!...
```

# 'sckit-learn' Kullanımı

```python
train = pd.read_csv('http://bit.ly/kaggletrain')
train.head()
```

```python
feature_cols=['Pclass','Parch']#Feature_cols dizisine kolon isimlerini attık onun aracılığıyla X adında yeni bir veriseti oluşturduk...
X=train.loc[:,feature_cols]
X.head()
```

```python
y=train.Survived
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()# Eğitim verilerine bir sınıflandırma modeli koyun...
logreg.fit(X,y)
```

```python
test = pd.read_csv('http://bit.ly/kaggletest')
test.head()
```

```python
X_new =test.loc[:,feature_cols]#Test verileri ile eşleşen bir özlellik matrisi yaratıyoruz...
X_new.shape
```

```python
new_pred_class = logreg.predict(X_new)#Test seti verilerine tahminşer yapmak için fitted model kullanın...
```

```python
passenger=pd.DataFrame({'Passenger Id':test.PassengerId,'Survived':new_pred_class})
#PassengerId için dataframe oluştur ve set tahminlerini dene...
```

```python
passenger.set_index('Passenger Id').head(12)#Passenge Id yi index olarak ayarlayarak ilk sütunun olduğundan emin olun...
```

```python
passenger.to_csv('sub_csv')#DataFrame i kaggle a gönderilebilcek bir csv dosyası haline getirin...
```

# Tekrarlayan Satırları Bulmak Ve Kaldırmak

```python
user_cols=['user_id','age','gender', 'occupation', 'zip_code']
users = pd.read_table('http://bit.ly/movieusers',sep='|' , header=None , names=user_cols,index_col='user_id' )
users.head()
```

```python
users.zip_code.duplicated().tail()#Yinelenen öğeleri bulur bir önceki öğe ile aynı ise öğe true farklı ise false içeren bir dizin döndürür...
```

```python
users.zip_code.duplicated().sum()#Tekarlayan öğe sayısını verir...
```

```python
users.duplicated().head()#Yinelenen satırları bulur...
```

```python
users.loc[users.duplicated(keep='first')]#İlk tekrarkamayı sayıma dahil etmez...
```

```python
users.loc[users.duplicated(keep='last')]#Son tekrarlamayı sayıma dahil etmez...
```

```python
users.loc[users.duplicated(keep=False),:]#Bütün tekrarlamaları true olarak işaretler...
```

```python
users.drop_duplicates(keep='first').shape#İlk tekrarlama hariç bütün takralayan satıraları drop eder...
```

```python
users.drop_duplicates(keep='last').shape#Son tekrarlama hariç bütün takralayan satıraları drop eder...
```

```python
users.drop_duplicates(keep=False).shape#Bütün tekaralamaları drop eder...
```

```python
users.duplicated(subset=['age','zip_code']).sum()#Belirtilen kolonlardaki takrarlamalrın toplamını verir...
```

```python
users.drop_duplicates(subset=['age','zip_code']).shape#Sadece belirtilen kolonlardaki tekrarlamaları drop eder...
```
